```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
import random

# 🧠 Step 1: Augmentation Functions
def jitter(x, sigma=0.03):
    return x + sigma * torch.randn_like(x)

def scaling(x, sigma=0.1):
    factor = torch.normal(1.0, sigma, size=(x.size(0), 1)).to(x.device)
    return x * factor

def augment_signal(x):
    return jitter(x) if random.random() > 0.5 else scaling(x)

# 🧠 Step 2: Dataset for EEG Contrastive Learning
class EEGContrastiveDatasetFromDF(Dataset):
    def __init__(self, df, column='conditioning_image'):
        self.df = df
        self.column = column

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        signal = np.array(self.df.iloc[idx][self.column], dtype=np.float32)  # shape: [128, 440]
        signal = torch.tensor(signal).float()  # shape: [128, 440]
        return augment_signal(signal.clone()), augment_signal(signal.clone())

# 🧠 Step 3: EEG Encoder (for 128-channel input)
class EEGEncoder(nn.Module):
    def __init__(self, in_channels=128, embedding_dim=512):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv1d(in_channels, 64, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.Conv1d(64, 128, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.AdaptiveAvgPool1d(1),
            nn.Flatten(),
            nn.Linear(128, embedding_dim)
        )

    def forward(self, x):  # x shape: [B, 128, 440]
        return self.encoder(x)

# 🧠 Step 4: NT-Xent Loss
def nt_xent_loss(z1, z2, temperature=0.5):
    z1, z2 = F.normalize(z1, dim=1), F.normalize(z2, dim=1)
    reps = torch.cat([z1, z2], dim=0)
    sim = torch.matmul(reps, reps.T)

    N = z1.size(0)
    mask = torch.eye(2 * N, dtype=torch.bool).to(z1.device)
    sim = sim.masked_fill(mask, float('-inf'))

    pos = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)], dim=0)
    pos = torch.exp(pos / temperature)
    denom = torch.exp(sim / temperature).sum(dim=1)
    loss = -torch.log(pos / denom)
    return loss.mean()

# 🧠 Step 5: SimCLR Training
def train_simclr(encoder, dataloader, epochs=10, lr=1e-4):
    encoder.train()
    encoder = encoder.to(device)
    optimizer = torch.optim.Adam(encoder.parameters(), lr=lr)

    for epoch in range(epochs):
        total_loss = 0
        for x1, x2 in dataloader:
            x1, x2 = x1.to(device), x2.to(device)
            z1 = encoder(x1)
            z2 = encoder(x2)
            loss = nt_xent_loss(z1, z2)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader):.4f}")

# 🧠 Step 6: Extract Embeddings
def extract_embeddings(df, encoder, column='conditioning_image'):
    encoder.eval()
    embeddings = []
    with torch.no_grad():
        for i in range(len(df)):
            signal = torch.tensor(np.array(df.iloc[i][column]), dtype=torch.float32).unsqueeze(0).to(device)  # [1, 128, 440]
            emb = encoder(signal).squeeze(0).cpu().numpy()
            embeddings.append(emb)
    return np.array(embeddings)

# ✅ Main Execution
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create dataset/dataloader
dataset = EEGContrastiveDatasetFromDF(df_train, column="conditioning_image")
dataloader = DataLoader(dataset, batch_size=512, shuffle=True)

# Initialize and train encoder
encoder = EEGEncoder(in_channels=128, embedding_dim=512)
train_simclr(encoder, dataloader, epochs=200)

# Extract and store embeddings
embeddings = extract_embeddings(df_train, encoder, column="conditioning_image")
df_train["eeg_embedding"] = embeddings.tolist()
```

## Raw EEG To Embeddings

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
import random

# 🧠 Step 1: Augmentation Functions
def jitter(x, sigma=0.03):
    return x + sigma * torch.randn_like(x)

def scaling(x, sigma=0.1):
    factor = torch.normal(1.0, sigma, size=(x.size(0), 1)).to(x.device)
    return x * factor

def augment_signal(x):
    return jitter(x) if random.random() > 0.5 else scaling(x)

# 🧠 Step 2: Positional Encoding
class PositionalEncoding(nn.Module):
    def __init__(self, dim, max_len=440):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, dim)
        position = torch.arange(0, max_len).unsqueeze(1).float()
        div_term = torch.exp(torch.arange(0, dim, 2).float() * -(np.log(10000.0) / dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(1, 2)  # Change here: Transpose to [1, dim, max_len]
        self.register_buffer('pe', pe)

    def forward(self, x):
        # Change here: Slice pe along dimension 2 to match x's shape
        return x + self.pe[:, :, :x.size(2)]

# 🧠 Step 3: Dataset for EEG Contrastive Learning
class EEGContrastiveDatasetFromDF(Dataset):
    def __init__(self, df, column='conditioning_image'):
        self.df = df
        self.column = column

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        signal = np.array(self.df.iloc[idx][self.column], dtype=np.float32)  # shape: [128, 440]
        signal = torch.tensor(signal).float()  # shape: [128, 440]
        return augment_signal(signal.clone()), augment_signal(signal.clone())

# 🧠 Step 4: EEG Encoder with Positional Encoding
class EEGEncoder(nn.Module):
    def __init__(self, in_channels=128, embedding_dim=512):
        super().__init__()
        self.positional_encoding = PositionalEncoding(in_channels, max_len=440) # Change here: Use in_channels for dim
        self.encoder = nn.Sequential(
            nn.Conv1d(in_channels, 64, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.Conv1d(64, 128, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.AdaptiveAvgPool1d(1),
            nn.Flatten(),
            nn.Linear(128, embedding_dim)
        )
        self.layer_norm = nn.LayerNorm(embedding_dim)

    def forward(self, x):  # x shape: [B, 128, 440]
        x = self.positional_encoding(x)
        x = self.encoder(x)
        return self.layer_norm(x)

# 🧠 Step 5: NT-Xent Loss
def nt_xent_loss(z1, z2, temperature=0.5):
    z1, z2 = F.normalize(z1, dim=1), F.normalize(z2, dim=1)
    reps = torch.cat([z1, z2], dim=0)
    sim = torch.matmul(reps, reps.T)

    N = z1.size(0)
    mask = torch.eye(2 * N, dtype=torch.bool).to(z1.device)
    sim = sim.masked_fill(mask, float('-inf'))

    pos = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)], dim=0)
    pos = torch.exp(pos / temperature)
    denom = torch.exp(sim / temperature).sum(dim=1)
    loss = -torch.log(pos / denom)
    return loss.mean()

# 🧠 Step 6: Mixup Augmentation (for contrastive learning)
def mixup(x1, x2, alpha=0.2):
    lambda_ = np.random.beta(alpha, alpha)
    mixed_x = lambda_ * x1 + (1 - lambda_) * x2
    return mixed_x

# 🧠 Step 7: SimCLR Training
def train_simclr(encoder, dataloader, epochs=10, lr=1e-4):
    encoder.train()
    encoder = encoder.to(device)
    optimizer = torch.optim.Adam(encoder.parameters(), lr=lr)

    for epoch in range(epochs):
        total_loss = 0
        for x1, x2 in dataloader:
            x1, x2 = x1.to(device), x2.to(device)

            # Applying Mixup augmentation (optional)
            mixed_x = mixup(x1, x2)  # This line is optional, can be added for additional regularization

            z1 = encoder(mixed_x)
            z2 = encoder(x2)  # Normal contrastive loss on mixed signal and original signal
            loss = nt_xent_loss(z1, z2)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader):.4f}")

# 🧠 Step 8: Extract Embeddings
def extract_embeddings(df, encoder, column='conditioning_image'):
    encoder.eval()
    embeddings = []
    with torch.no_grad():
        for i in range(len(df)):
            signal = torch.tensor(np.array(df.iloc[i][column]), dtype=torch.float32).unsqueeze(0).to(device)  # [1, 128, 440]
            emb = encoder(signal).squeeze(0).cpu().numpy()
            embeddings.append(emb)
    return np.array(embeddings)

# ✅ Main Execution
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Create dataset/dataloader
dataset = EEGContrastiveDatasetFromDF(df_train, column="conditioning_image")
dataloader = DataLoader(dataset, batch_size=512, shuffle=True)

# Initialize and train encoder
encoder = EEGEncoder(in_channels=128, embedding_dim=512)
train_simclr(encoder, dataloader, epochs=20)

# Extract and store embeddings
embeddings = extract_embeddings(df_train, encoder, column="conditioning_image")
df_train["eeg_embedding"] = embeddings.tolist()
```

```python
import torch
import torch.nn.functional as F
import numpy as np

def add_frequency_features(x, n_fft=64, hop_length=16):
    # x shape: [B, C, T]
    B, C, T = x.size()
    freq_features = []
    for i in range(C):
        # Compute STFT for each channel
        stft = torch.stft(x[:, i, :], n_fft=n_fft, hop_length=hop_length, return_complex=True)
        magnitude = torch.abs(stft)
        freq_features.append(magnitude)
    # Concatenate along the channel dimension
    freq_features = torch.cat(freq_features, dim=1)  # [B, C * (n_fft // 2 + 1), T // hop_length + 1]
    return freq_features

# Test the function
df_train["freq_features"] = df_train["conditioning_image"].apply(lambda x: add_frequency_features(torch.tensor(np.array(x)).unsqueeze(0)))
```
